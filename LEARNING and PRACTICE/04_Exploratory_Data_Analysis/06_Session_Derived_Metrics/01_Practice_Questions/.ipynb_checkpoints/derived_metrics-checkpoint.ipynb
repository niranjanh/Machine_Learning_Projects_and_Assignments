```python
import numpy as np
import pandas as pd
```

```python
odi_df = pd.read_csv("./odi-batting.csv")
odi_df.head(10)
```

### QUES: Which player scored the highest number of centuries?

### adding new column to determine if runs scored was century or not

```python
odi_df["century"] = odi_df["Runs"].apply(lambda runs: 1 if runs >= 100 else 0)
odi_df.head(10)
```

```python
odi_century_pivot_table = pd.pivot_table(odi_df, index="Player", values="century", aggfunc=np.sum)
odi_century_pivot_table
```

```python
odi_century_pivot_table["century"].max()
```

```python
odi_century_pivot_table[odi_century_pivot_table["century"] == 48]
```

### QUES:
#####  The ‘strike rate’ of a batsman is defined as the (number of runs scored / number of 100 balls faced). It is a measure of ‘how fast a batsman has played’.For example, if a batsman scores 40 runs in 50 balls, the strike rate is 80 runs/100 balls.

#### Among all the innings where a century was scored by a batsman, which one has the highest strike rate?

```python
odi_century_df = odi_df[odi_df["century"] == 1]
odi_century_df.head()
```

```python
odi_century_df["Strike Rate"] = (odi_century_df["Runs"] * 100) / odi_century_df["Balls"]
odi_century_df.head()
```

```python
odi_century_df["Strike Rate"].max()
```

```python
odi_century_df[odi_century_df["Strike Rate"] == 255.0]
```

### QUES:
#### In which year were the maximum number of centuries scored by Indian players?

```python
odi_df["year"] = pd.DatetimeIndex(odi_df["MatchDate"]).year
odi_df.head()
```

```python
odi_india_df = odi_df[odi_df["Country"] == "India"]
odi_india_df.head()
```

```python
century_year_pivot_table = pd.pivot_table(odi_india_df, index=["Player", "Country", "year"], values=["century"], aggfunc=np.sum)
century_year_pivot_table
```

```python
century_year_pivot_table["century"].max()
```

```python
century_year_pivot_table[century_year_pivot_table["century"] == 9]
```
