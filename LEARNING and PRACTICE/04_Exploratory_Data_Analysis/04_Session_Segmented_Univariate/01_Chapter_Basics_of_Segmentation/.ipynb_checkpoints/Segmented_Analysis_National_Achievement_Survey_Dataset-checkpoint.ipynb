## Segmented Analysis

### QUE 1: What is the impact of the variable 'Watch.TV' on Science marks?

```python
# So with pivot_table function we are telling pandas to group all students according to the different values in "Watch.TV" 
# and in that segment it will apply mean function on the values of "Science.." column
watch_tv_pivot_table = pd.pivot_table(nas_df, columns = ["Watch.TV"], values="Science..", aggfunc=np.mean)
watch_tv_pivot_table
```

### RESULT:
#### The Watch.TV variable doesn’t distinguish the students significantly in terms of Science marks
#### The average marks of students in Science in the all the labels are the same, i.e. around 36% marks.

### QUE 2:  What is the impact of the father’s education on the student’s Maths marks?

```python
father_education_pivot_table = pd.pivot_table(nas_df, columns = ["Mother.edu"], values="Maths..", aggfunc=np.mean)
father_education_pivot_table
```

### RESULT:
#### Students whose fathers have a degree and above get better marks on an average
#### Students whose fathers come under the label “degree & above” get the highest average marks among the labels, i.e. 34.87%

### QUE 3: What is the impact of 'Play.Games' on Reading marks?

```python
play_games_pivot_table = pd.pivot_table(nas_df, columns = ["Play.games"], values="Reading..", aggfunc=np.mean)
play_games_pivot_table
```

### RESULT:
#### Students who play games once a week score high in Reading on an average.
#### Students who play games once a week score 59.62% average marks in reading.
