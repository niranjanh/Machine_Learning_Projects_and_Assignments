Investment type analysis:

**Observation:**

Here I can see, some of the values in category list have 0 in them for ex. Alter0tive Medicine, A0lytics, 0notechnology and more.

Logically it seems like they should be Alternative Medicine, Analytics, Nanotechnology.

So changing these values replacing 0 with na.

```python
# function to replace "0" in category_list with na or NA
def replaceZeroWithna(text):
    if "0" in text:
        if text.startswith("0"):
            text = "Na" + text[1:]
        else:
            text = text[0:text.find("0")] + "na" + text[text.find("0") + 1:]
    
    return text
```

```python
mapping["category_list"] = mapping["category_list"].apply(replaceZeroWithna)
```

```python
# filter mapping_melted dataframe 
mapping_melted = mapping_melted[mapping_melted["value"] == 1]
mapping_melted.head()
```

```python
mapping_melted.shape
mapping_melted.notnull().sum()
```

```python
mapping_melted.reset_index()
# mapping_copy_melted["category_list_copy"] = mapping_copy_melted["category_list"]
b = mapping_melted["category_list"].tolist()
# b = []
a = top9["primary_sector"].tolist()
```

```python
# set category_list as the index of mapping_melted
mapping_melted.set_index("category_list", inplace = True)
```

```python
common = [value for value in a if value in b]
common
```

```python
def findTheParentSector(category):
    if category in common:
        return mapping_melted.loc[category, "Parent_sector"]
```
