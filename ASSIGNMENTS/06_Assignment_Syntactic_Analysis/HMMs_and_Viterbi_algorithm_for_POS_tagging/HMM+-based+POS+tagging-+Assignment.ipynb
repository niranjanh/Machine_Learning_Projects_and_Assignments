### Build the vanilla Viterbi based POS tagger

```python
# compute Emission Probability for a given word for a given tag

def word_given_tag(word, tag, train_bag=train_tagged_words):
    taglist = [pair for pair in train_bag if pair[1] == tag]
    
    # count of tag occurance
    tag_count = len(taglist)    
    
    w_in_tag = [pair[0] for pair in taglist if pair[0] == word]    
    
    # count of word occurance among tag occurance
    word_count_given_tag = len(w_in_tag)    
    
    return (word_count_given_tag, tag_count)
```

```python
# compute Transition Probabilities of a previous and next tag

def t2_given_t1(t2, t1, train_bag=train_tagged_words):
    
    # all tags
    tags = [pair[1] for pair in train_bag]
    
    # only tag t1
    t1_tags = [tag for tag in tags if tag == t1]
    
    # coutn of tag t1 occurance
    count_of_t1 = len(t1_tags)
    
    t2_given_t1 = [tags[index + 1] for index in range(len(tags) - 1) if tags[index] == t1 and tags[index + 1] == t2]
    
    # count of tag t2 occurance after tag t1
    count_t2_given_t1 = len(t2_given_t1)
    
    return(count_t2_given_t1, count_of_t1)
```

```python
# creating t x t transition matrix of tags
# each column is t2, each row is t1
# thus M(i, j) represents P(tj given ti)

tags_matrix = np.zeros((len(unique_tags), len(unique_tags)), dtype='float32')

for i, t1 in enumerate(list(unique_tags)):
    for j, t2 in enumerate(list(unique_tags)): 
        # computing transition prob
        tags_matrix[i, j] = t2_given_t1(t2, t1)[0]/t2_given_t1(t2, t1)[1]
```

### Viterbi Algorithm

For a given list of observations to be tagged, rather than computing the prob og all possible tag sequentially,
we assign the most likely tag to each word/observation using the only prev word.


Steps:

1. Given a sequence of words, Iterate through the sequence

2. For each word (starting from first word in sequence) calculate the product of emission probabilties and transition probabilties for all possible tags.

3. Assign the tag which has maximum probability obtained in step 2 above.

4. Move to the next word in sequence to repeat steps 2 and 3 above.

```python
# Vanilla Viterbi Algorithm
def Viterbi(words, train_bag = train_tagged_words):
    state = []
    
    T = list(set([pair[1] for pair in train_bag]))
    
    for key, word in enumerate(words):
        #initialise list of probability column for a given observation
        p = [] 
        for tag in T:
            if key == 0:
                transition_p = tags_df.loc['.', tag]
            else:
                transition_p = tags_df.loc[state[-1], tag]
                
            # compute emission and state probabilities
            emission_p = word_given_tag(words[key], tag)[0]/word_given_tag(words[key], tag)[1]
            state_probability = emission_p * transition_p    
            p.append(state_probability)
            
        pmax = max(p)
        # getting state for which probability is maximum
        state_max = T[p.index(pmax)] 
        state.append(state_max)
        
    return list(zip(words, state))
```

### Solve the problem of unknown words

### Method 1

#### Approach: 

assign based on transition probabilities only in case of unknown words as emission probability for unknown word is zero.

In addition to thta,

we  will apply weights based on the probability of tag occurance to the transition probabilities of tags and then use the resulting probability for predicting unknown words.

i.e. We will take into account that some POS tags are more likely to occur as compared to others.


```python
# use weighted transition probability of tags when emission probability is zero (in case of unknown words)
def Viterbi_method_1(words, train_bag = train_tagged_words):
    state = []
    T = list(set([pair[1] for pair in train_bag]))
    
    for key, word in enumerate(words):
        #initialise list of probability column for a given observation
        p = [] 
        p_transition =[] # list for storing transition probabilities
       
        for tag in T:
            if key == 0:
                transition_p = tags_df.loc['.', tag]
            else:
                transition_p = tags_df.loc[state[-1], tag]
                
            # compute emission and state probabilities
            emission_p = word_given_tag(words[key], tag)[0]/word_given_tag(words[key], tag)[1]
            state_probability = emission_p * transition_p    
            p.append(state_probability)
            
            # find POS tag occurance probability
            tag_p = [pair[1] for pair in all_tags_prob if pair[0] == tag ]
            
            # calculate the transition prob weighted by tag occurance probability.
            transition_p = tag_p[0]*transition_p             
            p_transition.append(transition_p)
            
        pmax = max(p)
        state_max = T[p.index(pmax)] 
        
      
        # if probability is zero (unknown word) then use weighted transition probability
        if(pmax==0):
            pmax = max(p_transition)
            state_max = T[p_transition.index(pmax)]                 
                           
        else:
            state_max = T[p.index(pmax)] 
        
        state.append(state_max)
    return list(zip(words, state))
```

## Method 2:

##### Approach:

We will modify the viterbi algorithm to use trigram tagger as a back-off tagger.

When the viterbi algorithm is not able to tag an unknown word, it will use the triagram-tagger as back off tagger.

```python
# Unigram Tagger
unigram_tagger = nltk.UnigramTagger(train_set)
ut_accuracy = unigram_tagger.evaluate(test_set)

print("Accuracy of the Unigram Tagger: ", ut_accuracy)
```

```python
# Rule Based Regex Tagger

# patterns for tagging using a rule based regex tagger -
patterns = [
    (r'^[aA-zZ].*[0-9]+','NOUN'),               # Alpha Numeric
    (r'.*ness$', 'NOUN'),
    (r'.*\'s$', 'NOUN'),                        # possessive nouns
    (r'.*s$', 'NOUN'),                          # plural nouns
    (r'.*', 'NOUN'),    
    (r'.*ly$', 'ADV'),
    (r'^(0|([*|-|$].*))','X'),                  # Any special character combination
    (r'.*ould$', 'X'),                          # modals
    (r'(The|the|A|a|An|an)$', 'DET'),
    (r'^([0-9]|[aA-zZ])+\-[aA-zZ]*$','ADJ'),
    (r'.*able$', 'ADJ'),                        # adjective like 100-megabytes 237-Seats
    (r'[aA-zZ]+(ed|ing|es)$', 'VERB'),          # Any word ending with 'ing' or 'ed' is a verb
    (r'[0-9].?[,\/]?[0-9]*','NUM')              # Numbers 
]
```

```python
rule_based_tagger = nltk.RegexpTagger(patterns)

# unigram tagger backed up by the rule-based tagger
rule_based_unigram_tagger = nltk.UnigramTagger(train_set, backoff = rule_based_tagger)

rbut_accuracy = rule_based_unigram_tagger.evaluate(test_set)

print("Accuracy of the Unigram Tagger backed up by the RegexpTagger: ", rbut_accuracy)
```

```python
# Bigram tagger

bigram_tagger = nltk.BigramTagger(train_set, backoff = rule_based_unigram_tagger)
bigram_tagger.evaluate(test_set)

bt_accuracy = bigram_tagger.evaluate(test_set)

print("Accuracy of the Bigram Tagger backed up by the rule_based_unigram_tagger: ", bt_accuracy)
```

```python
# Trigram tagger

trigram_tagger = nltk.TrigramTagger(train_set, backoff = bigram_tagger)
trigram_tagger.evaluate(test_set)

tt_accuracy = trigram_tagger.evaluate(test_set)

print("Accuracy of the Trigram Tagger backed up by the bigram_tagger: ", tt_accuracy)
```

```python
# A trigram tagger backed off by a rule based tagger.

def trigram_tagger_for_viterbi_algo(word, train_set = train_set):
    
    patterns = [
        (r'[aA-zZ]+(ed|ing|es)$', 'VERB'), # Any word ending with 'ing' or 'ed' is a verb
        (r'.*ly$', 'ADV'),
        (r'^([0-9]|[aA-zZ])+\-[aA-zZ]*$','ADJ'),
        (r'.*able$', 'ADJ'), 
        (r'.*ful$', 'ADJ'),
        (r'.*ous$', 'ADJ'),
        (r'^[aA-zZ].*[0-9]+','NOUN'),     # Alpha Numeric
        (r'.*ness$', 'NOUN'),
        (r'.*\'s$', 'NOUN'),              # possessive nouns - words ending with 's
        (r'.*s$', 'NOUN'),                # plural nouns
        (r'.*ers$', 'NOUN'),
        (r'.*ment$', 'NOUN'),
        (r'.*town$', 'NOUN'),
        (r'^(0|([*|-|$].*))','X'),        # Any special character combination
        (r'.*ould$', 'X'),
        (r'(The|the|A|a|An|an|That|that|This|this|Those|those|These|these)$', 'DET'), # That/this/these/those belong to the category of Demonstrative determiners
        (r'[0-9].?[,\/]?[0-9]*','NUM'),   # Numbers 
        (r'.*', 'NOUN')
    ]

    regex_based_tagger = nltk.RegexpTagger(patterns)

    # trigram backed up by the regex tagger
    trigram_regex_tagger = nltk.TrigramTagger(train_set, backoff = regex_based_tagger)
    
    return trigram_regex_tagger.tag_sents([[(word)]])
```

```python
def viterbi_method_2_with_trigram_tagger(words, train_bag = train_tagged_words):
    state = []
    tag_set = list(set([pair[1] for pair in train_bag]))
    
    for key, word in enumerate(words):
        if word not in vocab:
            # use the trigram tagger for unknown words.
            unknown_word_tag = trigram_tagger_for_viterbi_algo(word)
            
            for sent in unknown_word_tag:
                for tup in sent:
                    state.append(tup[1])
        else:            
            p = [] 
            for tag in tag_set:
                if key == 0:
                    transition_p = tags_df.loc['.', tag]
                else:
                    transition_p = tags_df.loc[state[-1], tag]
                
                # compute emission and state probabilities
                emission_p = word_given_tag(words[key], tag)[0]/word_given_tag(words[key], tag)[1]
                state_probability = emission_p * transition_p    
                p.append(state_probability)
            
            pmax = max(p)
            # getting state for which probability is maximum
            state_max = tag_set[p.index(pmax)] 
            state.append(state_max)
            
    return list(zip(words, state))
```
