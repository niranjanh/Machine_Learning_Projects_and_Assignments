### let us create a derived column from emp_length which will have experience of customer in int year.

```python
# [int(i) for i in test_string.split() if i.isdigit()]
def extractEmpExp(string):
    string = str(string)
    exp = 0
    for v in string.split(" "):
        if v.isdigit():
            exp = int(v)
            break
    
    return exp
```

```python
filtered_clean_loan["exp"] = loan['emp_length'].astype(str)
```

```python
filtered_clean_loan["exp"] = filtered_clean_loan["exp"].apply(extractEmpExp)
```

```python
filtered_clean_loan["exp"].value_counts()
```

```python
# lets see segmented analysis of loan_status acc to exp of employee
plt.figure(figsize=(15,5))
sns.boxplot(x=filtered_clean_loan["loan_status"], y=filtered_clean_loan["exp"].sort_values(ascending=False), data=filtered_clean_loan)
plt.show()
```
